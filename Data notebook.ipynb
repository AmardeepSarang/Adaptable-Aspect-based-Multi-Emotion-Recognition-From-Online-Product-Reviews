```python
import pandas as pd
import gzip
import json
```

```python


def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

```

```python
#get meta data for electronic products
toy_df = getDF('dataset/meta_Toys_and_Games.json.gz')
toy_df
```

```python
#get review data for electronic products
review_df = getDF('dataset/Toys_and_Games.json.gz')
review_df
```

```python
#pickle toy product, keep title, asin, and price
toy_df[["title","asin","price"]].to_pickle('dataset/meta_toy.pkl')
```

```python
%%time 
#get pickled toy reviews
unpickled_comp = pd.read_pickle('dataset/toys_reviews.pkl')
review_df=unpickled_comp
review_df
```

```python
rc_df=toy_df[toy_df['title'].str.contains('RC car |Remote control|remote control|car ')]
rc_df.shape[0]
```

```python
rc_df
```

```python
rc_reviews=pd.merge(review_df, rc_df, on='asin', how='inner')
rc_reviews
```

```python
review_count=rc_reviews.groupby(['asin','title']).reviewText.agg('count').to_frame('count').reset_index()
top_cars=review_count.sort_values(['count'],ascending=[False]).head(30)
top_cars.to_pickle('dataset/cars_asin_counts.pkl')
```

```python
top_cars
```

```python
top_car_reviews = rc_reviews.merge(top_cars["asin"])
top_car_reviews.to_pickle('dataset/top_car_reviews.pkl')
```
